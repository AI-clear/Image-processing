# 각 색상 생성

```python
import matplotlib.pyplot as plt
import numpy as np

image = np.zeros((200, 300, 3), dtype=np.uint8)

image[:100, :150] = [255, 0, 0]       
image[:100, 150:] = [0, 255, 0]       
image[100:, :150] = [0, 0, 255]       
image[100:, 150:] = [255, 255, 255]   

plt.imshow(image)
plt.axis('off')  
plt.show()
```

# matplot의 img을 이용하여 사진 읽고 출력하기

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

src= img.imread("img.png")
print(src.shape)
plt.imshow(src)
plt.show()
```

# 색감의 투명도 조절하기

```python
#alpha 채널은 투명도를 담당 각 채널은 (R,G,B,alpha)를 의미
import matplotlib
import numpy as np

h,w = src.shape[0], src.shape[1]

src_alpha = np.zeros((h,w,4),dtype=src.dtype)
src_alpha[:,:,:3] = src[:,:,:3]
alpha = np.zeros((h,w),dtype=src_alpha.dtype)

src_alpha[:,:,3] = alpha
plt.imshow(src_alpha)
plt.title('alpha = 0')
plt.show()

alpha += 0.5
src_alpha[:,:,3] = alpha
plt.imshow(src_alpha)
plt.title('alpha = 0.5')
plt.show()

alpha = 1
src_alpha[:,:,3] = alpha
plt.imshow(src_alpha)
plt.title('alpha = 1')
plt.show()
```

# 흰색 부분을 통해 background를 보이게 하는 img 

```python
alpha = 0
src_alpha[150:200,150:200,3] = alpha
plt.imshow(src_alpha)
plt.show()
plt.imsave("picture.png",src_alpha)
```

# CV2를 이용한 img 읽고 출력하기(BGR로 출력이 되서 색감이 이상해 보임)

```python
import cv2
src=cv2.imread("img.png")
print(src.shape, src.dtype, np.min(src), np.max(src))

# cv2.namedWindow('IMG')
# cv2.imshow('IMG',src)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(src)
plt.show()
```

# CV2 img convert BGR to RGB

```python
plt.imshow(np.flip(src,axis=2))
plt.show()

plt.imshow(src[:,:,::-1]) # 뒤에서 부터 읽음
plt.show()

src_RGB = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
plt.imshow(src_RGB)
plt.show()
```

# CH 단색 R,G,B으로 만 사진 출력
